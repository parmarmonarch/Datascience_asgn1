```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from random import randrange
import math
```

### Distance Function

```python
def distance_l2(x,y):
    sum_l2 = 0
    for i in range(len(x)):
        sum_l2+= (x[i]-y[i])**2
    return math.sqrt(sum_l2)
```

### Distance To Nearest Center


```python
def nearest_center_distance(x,G,P):
    dlist = []
    for i in G:
        dlist.append(distance_l2(P[i],P[x]))
    return min(dlist)
```

```python
def GreedyKCenter(k,dataset_name):
    
    ###### Read csv data ##########
    df = pd.read_csv(dataset_name)
    df = df.dropna()

    ###### Data Normalization ##########
    columns = list(df.columns)
    columns.pop(0)
    scaler = preprocessing.MinMaxScaler()
    x = df[columns].values
    scaled = scaler.fit_transform(x)
    df_temp = pd.DataFrame(scaled, columns=columns, index = df.index)
    df[columns] = df_temp

    G = []                                       # initialize list of centers
    distance = np.zeros(len(df))                 # initialize point distances
    G.append(randrange(len(df)))                 # randomly choose first center
    P = [df.iloc[x,1:] for x in range(len(df))]  # set of points

    for j in range(k-1):
        for i in range(len(df)):
            distance[i] = nearest_center_distance(i,G,P)
        G.append(distance.argmax())
        Delta = max(distance)
    return G, Delta
    
```

```python
GreedyKCenter(3,'cdata.csv')
```

## Optimal Solution for a subset of dataset for k=2, 3

```python
df = pd.read_csv('cdata.csv')
df_short = df.sample(n=10)
df_short = df_short.dropna()
```

```python
columns = list(df_short.columns)
columns.pop(0)
scaler = preprocessing.MinMaxScaler()
x = df_short[columns].values
scaled = scaler.fit_transform(x)
df_temp = pd.DataFrame(scaled, columns=columns, index = df_short.index)
df_short[columns] = df_temp
```

```python
G = []                                             # initialize list of centers
distance = np.zeros(len(df_short))                 # initialize point distances
P = [df_short.iloc[x,1:] for x in range(len(df_short))]
```

### Optimal clustering algo for k=2

```python
out = []
for i in range(len(df_short)):   
    G.append(i)
    for j in range(len(df_short)):
        if i!=j:
            G.append(j)
            for h in range(len(df_short)):
                if h!=i and h!=j:
                    distance[h] = nearest_center_distance(h,G,P)
            out.append(max(distance))
            G=[]
deltaopt2 = min(out)
```

```python
df_short.to_csv('temp2',index=False)
```

```python
DeltaGreedy2 = [GreedyKCenter(2,'temp2')][0][-1]
```

```python
approximation_factor2 = DeltaGreedy2/deltaopt2
```

```python
print("approximation factor for k=2 is ", approximation_factor2)
```

### Optimal clustering algo for k=3


```python
G = []                                             # initialize list of centers
distance = np.zeros(len(df_short))   
out = []
for l in range(len(df_short)):
    G.append(l)
    for i in range(len(df_short)):
        if i!=l:
            G.append(i)
            for j in range(len(df_short)):
                if j!=i and j!=l:
                    G.append(j)
                    for h in range(len(df_short)):
                        if h!=i and h!=j and h!=l:
                            distance[h] = nearest_center_distance(h,G,P)
                    out.append(max(distance))
                    G=[]
deltaopt3 = min(out)
```

```python
df_short.to_csv('temp3',index=False)
```

```python
DeltaGreedy3 = [GreedyKCenter(3,'temp3')][0][-1]
```

```python
approximation_factor3 = DeltaGreedy3/deltaopt3
```

```python
print("approximation factor for k=3 is ", approximation_factor3)
```
